# Matplotlib

* Most popular plotting library for python <br>
* Gives control over every aspect of a figure <br>
* Feels similar to MatLab

> conda install matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

### Functional

```python
plt.plot(y,x, 'p-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')

plt.subplot(2,1,2)
plt.plot(y,x,'b')
```

# Object orientated 

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0.8,0.8])
axes1.set_title("LARGER PLOT")
axes1.set_xlabel("Number of high quality experiments")
axes1.set_ylabel("Chances of success")
axes1.plot(x,y)
axes2 = fig.add_axes([0.1,0.4,0.4,0.3])
axes2.set_title("SMALLER PLOT")
axes2.set_xlabel("Chances of ending average lives")
axes2.plot(y,x)
```

first: starting x <br>
second: starting y <br>
third: width <br>
fourth: height <br>

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(y,x)
axes[1].plot(x,y)

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(12,2))
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=800)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Squared")
ax.plot(x,x**3,label="X Cubed")

ax.legend()
```

## Graph appearance

color <br>
lw -> line width <br>
alpha -> transparency <br>
linestyle -> ls ('--', '-.', ':', 'steps') <br>
marker ('1','o','+') <br>
markersize <br>
markerfacecolor <br>
markerfacecolor <br>
markeredgewidth <br>
markeredgecolor <br>

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#810034', lw=3, ls='-.')
```
